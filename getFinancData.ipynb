# **Dados financeiros**

```python
import pandas_datareader.data as web
import datetime as dt
```

## Obtendo dados 

```python
#data de início
start = dt.datetime(2021,1,1)
#data de fim
end = dt.datetime.now()

#quais ações deseja-se analisar
tickers = ['PETR4.SA','VALE3.SA']

df=web.get_data_yahoo(tickers,start,end)
```

```python
df.head()
```

## Visualizando os ativos com *candlestick plot*

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
candlestick = go.Figure(data = [go.Candlestick(x = df.index, 
                                               open = df[('Open','PETR4.SA')], 
                                               high = df[('High','PETR4.SA')], 
                                               low = df[('Low','PETR4.SA')], 
                                               close = df[('Close','PETR4.SA')])])

candlestick.update_layout(xaxis_rangeslider_visible = False, title = 'Petrobras')
candlestick.update_xaxes(title_text = 'Date')
candlestick.update_yaxes(title_text = 'Fechamento', tickprefix = "R$")

candlestick.show()
```
